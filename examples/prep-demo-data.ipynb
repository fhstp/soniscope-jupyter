```python
import pandas as pd
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00275/Bike-Sharing-Dataset.zip'
zip_file_name = 'Bike-Sharing-Dataset.zip'
csv_file = 'day.csv'
```

```python
# load data file from zip archive on UCI ML 
# hack based on <https://stackoverflow.com/a/67298378/1140589>
import zipfile
from urllib.request import urlopen
import shutil
import os

# extracting zipfile from URL
with urlopen(url) as response, open(zip_file_name, 'wb') as out_file:
    shutil.copyfileobj(response, out_file)

    # extracting required file from zipfile
    with zipfile.ZipFile(zip_file_name) as zf:
        zf.extract(csv_file)

# deleting the zipfile from the directory
os.remove(zip_file_name)
```

```python
# loading data from the file
daily = pd.read_csv(csv_file)
daily.head(3)
```

```python
daily2 = daily.copy()
daily2['temp °C'] = daily['temp'] * (39 + 8) - 8
daily2['atemp °C'] = daily['atemp'] * (50 + 16) - 16
daily2['windspeed'] = daily['windspeed'] * 67
daily2['hum %'] = daily['hum'] * 100
daily2['year'] = daily['yr'] + 2011
```

```python
# daily2['windspeed'] = daily['windspeed'] * 67
daily2.loc[daily['weathersit'] == 1, 'weathersit'] = 'Clear, Few clouds, Partly cloudy, Partly cloudy'
daily2.loc[daily['weathersit'] == 2, 'weathersit'] = 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist'
daily2.loc[daily['weathersit'] == 3, 'weathersit'] = 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds'
daily2.loc[daily['weathersit'] == 4, 'weathersit'] = 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'
```

```python
daily2.loc[daily['season'] == 1, 'season'] = 'winter'
daily2.loc[daily['season'] == 2, 'season'] = 'spring'
daily2.loc[daily['season'] == 3, 'season'] = 'summer'
daily2.loc[daily['season'] == 4, 'season'] = 'fall'
```

```python
daily2.loc[daily['weekday'] == 0, 'weekday'] = 'Sunday'
daily2.loc[daily['weekday'] == 1, 'weekday'] = 'Monday'
daily2.loc[daily['weekday'] == 2, 'weekday'] = 'Tuesday'
daily2.loc[daily['weekday'] == 3, 'weekday'] = 'Wednesday'
daily2.loc[daily['weekday'] == 4, 'weekday'] = 'Thursday'
daily2.loc[daily['weekday'] == 5, 'weekday'] = 'Friday'
daily2.loc[daily['weekday'] == 6, 'weekday'] = 'Saturday'
```

```python
daily2 = daily2.drop(['temp', 'atemp', 'hum', 'yr'], axis=1)
```
